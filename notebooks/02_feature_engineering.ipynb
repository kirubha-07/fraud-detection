# 02 — Feature Engineering

Every feature below has a **business rationale** documented before we test whether it predicts fraud.

| Feature | Business Rationale |
|---|---|
| `errorBalanceOrig` | Fraudulent transactions often fail to debit the origin account correctly |
| `errorBalanceDest` | Destination balance may not match expected credit |
| `amountToOldBalanceRatio` | Draining full balance (ratio ≈ 1) signals account takeover |
| `drainsFullBalance` | Binary flag for near-total balance extraction |
| `isNewDestAccount` | Fraudsters route to zero-balance (new) destination accounts |
| `type_*` one-hot | Transaction type determines fraud eligibility and behavior |
